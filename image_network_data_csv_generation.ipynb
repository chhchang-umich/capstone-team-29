```python
file_path = '/content/drive/My Drive/Capstone Project/NodeXL_tweets_labels.csv'
df=pd.read_csv(file_path)
```

```python
df.shape
```

```python
df.head()
```

```python
image_url_labels_dict = pd.Series(list(zip(df['Image URL'], df['Labels'])), index=df['Image Name']).to_dict()

print(image_url_labels_dict)
```

```python
# empty list to hold the results
results = []

# iterate through the image pairs and compute the number of shared labels
for image1, image2 in combinations(image_url_labels_dict.keys(), 2):
  labels1 = eval(image_url_labels_dict[image1][1])
  labels2 = eval(image_url_labels_dict[image2][1])
  shared_labels = set(labels1) & set(labels2)

  if shared_labels:
    weight = len(shared_labels)
    # save only when weight is over 1 shared label
    if weight > 1:
      results.append({
          'image_a_name': image1,
          'image_a_url': image_url_labels_dict[image1][0],
          'image_b_name': image2,
          'image_b_url': image_url_labels_dict[image2][0],
          'weight (number of shared labels)': weight,
          'shared_labels': list(shared_labels)
          })

# Convert the results to DataFrame
image_network_df = pd.DataFrame(results)
```

```python
# Save the df as CSV
image_network_df.to_csv('/content/drive/My Drive/Capstone Project/image_pairs_shared_labels.v1.csv', index=False)

print("The csv file is created.")
```
